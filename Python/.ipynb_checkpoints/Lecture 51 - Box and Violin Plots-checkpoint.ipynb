```python
import numpy as np
import pandas as pd
from numpy.random import randn

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data1 = randn(1e2)
data2 = randn(1e2)
```

```python
sns.boxplot([data1, data2])
```

```python
# Para incluir todos los outliers y que los bigotes representen el rango

sns.boxplot([data1, data2], whis = np.inf)
```

```python
sns.boxplot(data1, vert = False)
```

```python
# Normal Dist

data1 = stats.norm(0, 5).rvs(100)

# Two gamma distribution and cocatenate

data2 = np.concatenate([stats.gamma(5).rvs(50) - 1,
                       -1 * stats.gamma(5).rvs(50)])
# Boxplot both datasets

sns.boxplot([data1, data2], whis = np.inf)
```

```python
sns.violinplot([data1, data2])
```

```python
sns.violinplot([data1, data2], bw = 0.01)
```

```python
sns.violinplot([data1, data2], inner = "stick")
```
