```python
import numpy as np
import pandas as pd
from numpy.random import randn

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.lmplot("total_bill", "tip", tips)
```

```python
sns.lmplot("total_bill", "tip", tips,
           scatter_kws = {"marker": "+", "color": "black"},
           line_kws = {"color": "red", "lw": 2})
```

```python
sns.lmplot("total_bill", "tip", tips,
           scatter_kws = {"marker": "+", "color": "black"},
           line_kws = {"color": "red", "lw": 2},
           order = 4)
```

```python
sns.lmplot("total_bill", "tip", tips, fit_reg = False)
```

```python
tips.head()
```

```python
tips["%tip"] = 100 * tips.tip / tips.total_bill
tips.head()
```

```python
sns.lmplot("size", "%tip", tips, x_jitter = 0.2)
```

```python
sns.lmplot("size", "%tip", tips, x_estimator = np.mean)
```

```python
sns.lmplot(x = "total_bill", y = "%tip", data = tips, hue = "day")
```

```python
sns.lmplot("total_bill", "%tip", tips, lowess = True,
           line_kws = {"color": "red"})
```
