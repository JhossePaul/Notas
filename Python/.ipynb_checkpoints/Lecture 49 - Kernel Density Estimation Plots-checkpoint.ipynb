```python
import pandas as pd
import numpy as np
from scipy import stats
from numpy.random import randn
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data1 = randn(25)
data2 = randn(100)
```

```python
sns.rugplot(data1)
plt.ylim(0, 1)
```

```python
plt.hist(data1, alpha = 0.5)
sns.rugplot(data1)
```

```python
sns.rugplot(data1)

x_min = data1.min() - 2
x_max = data1.max() + 2

x_axis = np.linspace(x_min, x_max, 100)

bandwidth = ( (4 * data1.std() ** 5) / (3 * len(data1)) ) ** 0.2
kernel_list = []

for data_point in data1:
        # Creating a kernel for each point and append it to the kernel_list
        kernel = stats.norm(data_point, bandwidth).pdf(x_axis)
        kernel_list.append(kernel)
        
        # Scale for plotting
        kernel  = kernel / kernel.max() * 0.4
        
        plt.plot(x_axis, kernel, color = "gray", alpha = 0.5)
    
plt.ylim(0, 1)
```

```python
sum_of_kde = np.sum(kernel_list, axis = 0)
fig = plt.plot(x_axis, sum_of_kde, color = "red")
sns.rugplot(data1)

plt.yticks([])

plt.suptitle("Sum of KDE basis functions")
```

```python
sns.kdeplot(data1)
```

```python
sns.rugplot(data1, color = "black")

for bw in np.arange(0.5, 2, 0.25):
    sns.kdeplot(data1, bw = bw, lw = 1.8, label = bw)
```

```python
kernel_options = ["biw", "cos", "epa", "gau", "tri", "triw"]

for kern in kernel_options:
    sns.kdeplot(data1, kernel = kern, label = kern, shade = True)
```

```python
sns.kdeplot(data1, vertical = True)
```

```python
sns.kdeplot(data1, cumulative = True)
```

```python
mean = [0, 0]

cov = [[1, 0], [0, 100]]

data2 = np.random.multivariate_normal(mean, cov, 1000)

dframe = pd.DataFrame(data2, columns = ["X", "Y"])

print dframe.head()

sns.kdeplot(dframe)
```

```python
sns.kdeplot(dframe.X, dframe.Y, shade = True)
```

```python
sns.kdeplot(dframe.X, dframe.Y, shade = True, bw = 1)
```

```python
sns.kdeplot(dframe, bw = "silverman")
```

```python
sns.jointplot("X", "Y", dframe, kind = "kde")
```

```python
sns.jointplot("X", "Y", dframe, kind = "hex")
```
