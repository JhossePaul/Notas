```python
import numpy as np
import pandas as pd
from numpy.random import randn

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
dataset = randn(1e2)
```

```python
sns.distplot(dataset, bins = 25)
```

```python
sns.distplot(dataset, rug = True, hist = False)
```

```python
sns.distplot(dataset, bins = 25, rug = True,
             kde_kws = {"color": "red", "label": "KDE Plot"},
             hist_kws = {"alpha": 0.5, "color": "orange", "label": "Histogram"},
             rug_kws = {"color": "black"})
```

```python
from pandas import Series

my_series = Series(dataset, name = "My Data")
my_series
```

```python
sns.distplot(my_series, bins = 25)
```
